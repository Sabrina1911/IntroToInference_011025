### Recap: Maintaining the Instructor vs Student Branch Workflow

During our development session, we went through the full lifecycle of setting up and maintaining
a branching workflow for this project. Hereâ€™s a summary of the key steps we completed together:

- **Initial Setup**: We created the `main` branch for students and the `instructor` branch for teaching staff.
  The `main` branch only contains challenge/solution notebooks that students should see, while
  the `instructor` branch contains additional materials such as the full source notebook,
  utility scripts, configuration files, and documentation.

- **Switching Between Branches**: We practiced moving between `main` and `instructor` branches using commands like:
  ```bash
  git checkout main
  git checkout instructor
  ```
  This workflow ensures that edits to the source notebook and utilities are isolated in the instructor branch.

- **Protecting Student Branch**: We updated `.gitignore` so that sensitive or instructor-only files
  (such as `utils/`, `documentation/`, `config.json`, and `IntroToDataStructures_Workshop.ipynb`) are never committed to `main`.

- **Accidental File Deletion and Recovery**: At one point, instructor-only files were mistakenly removed from the branch.
  We restored them by checking out the files from the `instructor` branch using commands like:
  ```bash
  git checkout instructor -- utils/ documentation/ config.json SETUP.md IntroToDataStructures_Workshop.ipynb
  ```
  This highlighted the importance of keeping backups in the `instructor` branch.

- **Updating Workshop Files**: When changes were made to the source notebook (`IntroToDataStructures_Workshop.ipynb`),
  we used the `toggle_solutions.py` utility to generate the correct student-facing versions:
  - `IntroToDataStructures_Workshop_challenge.ipynb` before/during the workshop
  - `IntroToDataStructures_Workshop_solutions.ipynb` after the workshop

- **Synchronizing Branches**: We ensured that updates from the source notebook in `instructor`
  flowed correctly into the challenge/solution notebooks in `main`, without exposing instructor-only files.

Together, these steps form a robust workflow where students always receive clean materials from the `main` branch,
while instructors retain full control of the source and teaching utilities in the `instructor` branch.
